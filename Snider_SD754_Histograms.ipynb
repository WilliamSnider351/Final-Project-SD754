```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.ticker as mtick

# Model names and datasets
models = [
    'CNN3_MNIST', 'LRCNN2_MNIST',
    'CNN3_FashionMNIST', 'LRCNN2_FashionMNIST',
    'CNN3_CIFAR10', 'CNN6_CIFAR10',
    'VOTE3CNN_CIFAR10', 'RESNET8_CIFAR10', 'WRN2810_CIFAR10',
    'CNN5_CIFAR100'
]

datasets = [
    'MNIST Digits', 'MNIST Digits',
    'Fashion MNIST', 'Fashion MNIST',
    'CIFAR-10', 'CIFAR-10', 'CIFAR-10', 'CIFAR-10', 'CIFAR-10',
    'CIFAR-100'
]

# Updated accuracy values (converted to percent)
accuracy = [
    0.9917,  # CNN3_MNIST
    0.9926,  # LRCNN2_MNIST
    0.9230,  # CNN3_FashionMNIST
    0.9205,  # LRCNN2_FashionMNIST
    0.8094,  # CNN3_CIFAR10
    0.8653,  # CNN6_CIFAR10
    0.8046,  # VOTE3CNN_CIFAR10
    0.8187,  # RESNET8_CIFAR10
    0.7583,  # WRN2810_CIFAR10
    0.5116   # CNN5_CIFAR100 ‚Üê updated value
]

accuracy_percent = [val * 100 for val in accuracy]

# Create DataFrame
df = pd.DataFrame({
    'Model': models,
    'Dataset': datasets,
    'Percent correct': accuracy_percent
})

# Color palette (each bar a different color)
palette = sns.color_palette("hsv", len(models))

# Plot
plt.figure(figsize=(16, 6))
sns.barplot(data=df, x='Model', y='Percent correct', palette=palette)

# Style
plt.title('Classification Performance')
plt.xlabel('Model')
plt.ylabel('Percent correct')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add vertical separators between dataset groups
group_change_indices = [i for i in range(1, len(datasets)) if datasets[i] != datasets[i-1]]
for idx in group_change_indices:
    plt.axvline(x=idx - 0.5, color='gray', linestyle='--', alpha=0.5)

# Add dataset labels above the bars
group_centers = {
    'MNIST Digits': 0.5,
    'Fashion MNIST': 2.5,
    'CIFAR-10': 6,
    'CIFAR-100': 9
}
for dataset, x in group_centers.items():
    plt.text(x, max(accuracy_percent) + 1, dataset, ha='center', fontsize=11, fontweight='bold')

# Format Y-axis labels with 2 decimal places and a %
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:.2f}%'))

plt.tight_layout()
plt.show()
```
